```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# use for reference job id 6453519
SNN = np.load("./shallow_out/6453519/ROC.pdfNN_classifier.np.npy")
cMVA = np.load("./shallow_out/6453519/ROC.pdfJet_cMVA.np.npy")
RNN = np.load("./RNN_out/lstm64/ROC.npy")
#RNNb = np.load("./RNN_out/local2/RNN_ROC-50-large.npy")
```

```python
fig = plt.figure(figsize=(10,6))
plt.plot(SNN[1,:], SNN[0,:], label = "shallow network")
plt.plot(RNN[1,:], RNN[0,:], label = "LSTM-64")
#plt.plot(RNNb[1,:], RNNb[0,:], label = "LSTM-64")
plt.plot(cMVA[1,:], cMVA[0,:], label = "cMVA")
plt.yscale('log')
axes = plt.gca()
axes.set_ylim([1e-2,1])
plt.legend(loc = "upper left")
plt.xlabel('b jet efficiency')
plt.ylabel('misidentification prob.')
plt.show()
```
